# Summarization

Here I want to visually inspect the output from summarizing the abstracts

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

DIR_NAME = '/home/khev/2TB-harddrive/data/openNMT-summarization/arxiv/'
f_yhat = os.path.join(DIR_NAME, 'test/predicted.title.txt')
f_y = os.path.join(DIR_NAME, 'test/test.title.txt')
f_x = os.path.join(DIR_NAME,'test/test.abstract.txt')

#Load y's
df1 = pd.read_csv(f_x,sep='\n',header=None)
df2 = pd.read_csv(f_y,sep='\n',header=None)
df3 = pd.read_csv(f_yhat,sep='\n', header=None)
```

```python
df = pd.concat([df1,df2,df3], axis=1)
df.columns = ['abstract','title true','title predicted']
row = df.iloc[0]
```

```python
for i in range(10000):
    row = df.iloc[i]
    print('Abstract: {} \n'.format(row[0]))
    print('Title true: {} \n'.format(row[1]))
    print('Title predicted: {} \n \n'.format(row[2]))
```
